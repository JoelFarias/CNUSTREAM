## Limpando a base

### Ameaçados

```python
ameacados = pd.read_excel("CPT/CPT - excel/ameacados-de-morte-2022.xlsx")
ameacados.sample(5)
```

```python
filters = ["Subtotal", "Nordeste", "Norte", "Sudeste", "Sul", "Centro-Oeste", "Municípios", "Total", "Brasil", 'Acre', 'Alagoas', 'Amapá', 'Amazonas', 'Bahia', 'Ceará', 'Distrito Federal', 'Espírito Santo', 'Goiás', 'Maranhão', 'Mato Grosso', 'Mato Grosso do Sul', 'Minas Gerais', 'Pará', 'Paraíba', 'Paraná', 'Pernambuco', 'Piauí', 'Rio de Janeiro', 'Rio Grande do Norte', 'Rio Grande do Sul', 'Rondônia', 'Roraima', 'Santa Catarina', 'São Paulo', 'Sergipe', 'Tocantins']
ameacados_filter = ameacados[~ameacados.UF.isin(filters)]
ameacados_filter
```

### Áreas em conflitos

```python
areas_conflito = pd.read_excel("CPT/CPT - excel/areas-em-conflito-2022.xlsx")
areas_conflito_filter = areas_conflito.dropna(subset=["Nome do Conflito", "Municípios"], how="any")

areas_conflito_filter
```

### Assasinatos

```python
assasinatos = pd.read_excel("CPT/CPT - excel/assassinatos - 2022.xlsx", header=1)
assasinatos_filter = assasinatos[~assasinatos["Municípios"].isin(filters)]
assasinatos_filter = assasinatos_filter[assasinatos_filter["Municípios"].notna()]
assasinatos_filter
```

### Conflitos

```python
conflitos = pd.read_excel("CPT/CPT - excel/conflitos-por-terra-2022.xlsx", header=35)
conflitos_filter = conflitos.drop(["Unnamed: 2", "Unnamed: 4", "Unnamed: 6", "Unnamed: 8"], axis=1)
conflitos_filter = conflitos_filter.dropna(subset="Município")

conflitos_filter
```

### Ocupação

```python
ocupacao = pd.read_excel("CPT/CPT - excel/ocupacoes-retomadas-2022.xlsx", header=1)
ocupacao_filter = ocupacao[~ocupacao["Municípios"].isin(filters)]
ocupacao_filter = ocupacao_filter.dropna(subset="Nome do Imóvel")
ocupacao_filter
```

### Tentativas de assassinato

```python
tentativas = pd.read_excel("CPT/CPT - excel/tentativas-de-assassinatos-2022.xlsx", header=1)
tentativas_filter = tentativas[~tentativas["Municípios"].isin(filters)]
tentativas_filter = tentativas_filter.dropna(subset="Nome do Conflito")
tentativas_filter
```

### Trabalho escravo

```python
escravo = pd.read_excel("CPT/CPT - excel/trabalho-escravo-2019.xlsx", header=2)
escravo_filter = escravo[~escravo["Municipios"].isin(filters)]
escravo_filter = escravo_filter.dropna(subset="Municipios").drop("Unnamed: 0", axis=1)
escravo_filter
```

## Salvando arquivos

```python
with pd.ExcelWriter("CPT/CPT.xlsx", mode="w", engine="openpyxl") as writer:
    ameacados_filter.to_excel(writer, sheet_name="Ameaçados", index=False)  
    areas_conflito_filter.to_excel(writer, sheet_name="Áreas em Conflito", index=False)
    assasinatos_filter.to_excel(writer, sheet_name="Assassinatos", index=False)
    conflitos_filter.to_excel(writer, sheet_name="Conflitos por Terra", index=False)
    ocupacao_filter.to_excel(writer, sheet_name="Ocupações Retomadas", index=False)
    tentativas_filter.to_excel(writer, sheet_name="Tentativas de Assassinatos", index=False)
    escravo_filter.to_excel(writer, sheet_name="Trabalho Escravo", index=False) 
```

## Pegando informações do CPT

```python
import unidecode
import pandas as pd
import geopandas as gpd


def listar_mun(column, sep):  
    muns = set()
    for i in column.__iter__():
        if i.find(sep) != -1:
            i = i.split(sep)
            for j in i:
                muns.add(j)
        else:
            muns.add(i)
    return muns

def remover_acentos(texto):
    return unidecode.unidecode(texto)
```

```python
cnuc = gpd.read_file("areas-PA/áreas-selecionadas/cnuc/cnuc.shp")
cnuc
```

Pegando os municípios

```python
muns = listar_mun(cnuc["municipio"], ' , ')
muns
```

### Ameaçados

```python
ameacados = pd.read_excel("CPT/CPT.xlsx", sheet_name="Ameaçados")
ameacados[ameacados["UF"] == "PA"]
```

### Áreas em conflito

```python
areas_conflito = pd.read_excel("CPT/CPT.xlsx", sheet_name="Áreas em Conflito")
areas_conflito["mun"] = areas_conflito["Municípios"].apply(remover_acentos)
areas_conflito["mun"] = areas_conflito["mun"].str.lower()

areas_conflito[areas_conflito["mun"].isin(muns)]
```

```python
areas_conflito_filter = pd.DataFrame(columns=areas_conflito.columns)
for mun in muns:
    areas_conflito_filter = pd.concat([areas_conflito_filter, areas_conflito[areas_conflito["mun"].str.contains(mun)]])

areas_conflito_filter
```

### Assasinatos

```python
assassinatos =  pd.read_excel("CPT/CPT.xlsx", sheet_name="Assassinatos")
assassinatos["mun"] = assassinatos["Municípios"].str.lower()
assassinatos["mun"] = assassinatos["mun"].apply(remover_acentos)


assassinatos[assassinatos["mun"].isin(muns)]
```

### Conflitos por Terra

```python
def separar_estado(mun):
    return mun.split("(")[0]

conflitos = pd.read_excel("CPT/CPT.xlsx", sheet_name="Conflitos por Terra")
conflitos["mun"] = conflitos["Município"].apply(separar_estado)
conflitos["mun"] = conflitos["mun"].str.lower()
conflitos["mun"] = conflitos["mun"].apply(remover_acentos)

conflitos[conflitos["mun"].isin(muns)]
```

### Ocupação

```python
ocupacao = pd.read_excel("CPT/CPT.xlsx", sheet_name="Ocupações Retomadas")
ocupacao["mun"] = ocupacao["Municípios"].apply(remover_acentos)
ocupacao["mun"] = ocupacao["mun"].str.lower()

ocupacao_filter = pd.DataFrame(columns=ocupacao.columns)
for mun in muns:
    ocupacao_filter = pd.concat([ocupacao_filter, ocupacao[ocupacao["mun"].str.contains(mun)]])

ocupacao_filter
```

### Tentativas de assasinato

```python
tentativas =  pd.read_excel("CPT/CPT.xlsx", sheet_name="Tentativas de Assassinatos")
tentativas["mun"] = tentativas["Municípios"].str.lower()
tentativas["mun"] = tentativas["mun"].apply(remover_acentos)


tentativas[tentativas["mun"].isin(muns)]
```
